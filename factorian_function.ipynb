```python
def print_no(n):
    if n ==0 :
        return
    print(n,end=" ")
    print_no(n + 1)
print_no(3) 
    
```

```python
def fact(i):
    if i == 1:
        return 1
    else:
        return i * fact(i - 1)
    

i =int(input("Enter number"))
fact(i)
```

```python
#fibonacci
def fact(i):
    if i == 1:
        return 1
    else:
        return fact(i - 1) + fact(i - 2)
    print(i)
i =int(input("Enter number"))
fact(i)
```
