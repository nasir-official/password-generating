```python
#Admin or Student Verification system
pas=1234
username="admin"
user=input("what is your role")
if user=="admin":
    q=int(input("enter your password"))
    uname=input("enter your user name")
    if q==pas and uname==username:
        print("your admin")
    else:
        print("your not admin")   
elif user=="student":
    s=int(input("enter your password"))
    if s==pas:
        print("your student")
    else:
        print("your not student")
else:
    print("your not part of system")
```
