```python
#=========HOSPITAL MANAGEMENT SYSTEM========
#--------Base Class
class Person:
    def __init__(self, name, age): #conxtractor
        self.name = name
        self.age = age

    def details(self):
        return f"Name: {self.name}    |    Age: {self.age}"

class Patient(Person):  #inheritance
    def __init__(self, patient_id, name, age, disease):
        super().__init__(name, age)
        self.__patient_id = patient_id   # Encapsulation
        self.disease = disease

    def patient_id(self):
        return self.__patient_id

    def details(self):
        return f"Patient ID: {self.__patient_id}    |   {super().details()}   |    Disease: {self.disease}"

    def __str__(self):
        return self.details()

class Doctor(Person):
    def __init__(self, doctor_id, name, age, specialization):
        super().__init__(name, age)
        self.doctor_id = doctor_id
        self.specialization = specialization

    def details(self):
        return f"Doctor ID: {self.doctor_id}   |   {super().details()}    |    Specialization: {self.specialization}"

    def __str__(self):
        return self.details()

class Appointment:
    def __init__(self, patient, doctor, date):
        self.patient = patient
        self.doctor = doctor
        self.date = date

    def __str__(self):
        return f"Appointment Date: {self.date}   |    Patient: {self.patient.name}   |    Doctor: {self.doctor.name}"

class Billing:
    def __init__(self, patient, days, dailycharge):
        self.patient = patient
        self.days = days
        self.dailycharge = dailycharge

    def calculatebill(self):
        return self.days * self.dailycharge

    def __str__(self):
        return f"Patient: {self.patient.name}   |    Total Bill: Rs.{self.calculatebill()}"

class info:
    def __init__(self, name):
        self.name = name
        self.patients = []
        self.doctors = []
        self.appointments = []
        print(f"========THE {self.name}==========")

    def addpatient(self, patient):
        self.patients.append(patient)

    def adddoctor(self, doctor):
        self.doctors.append(doctor)

    def scheduleappointment(self, appointment):
        self.appointments.append(appointment)

    def showpatients(self):
        print("\n--- Patient Records ---")
        for patient in self.patients:
            print(patient)

    def showdoctors(self):
        print("\n--- Doctor Records ---")
        for doctor in self.doctors:
            print(doctor)

    def showappointments(self):
        print("\n--- Appointments ---")
        for appointment in self.appointments:
            print(appointment)

hospital = info("City Hospital")
# Adding Patients
p1 = Patient(101, "Ali", 30, "Fever")
p2 = Patient(102, "Sara", 65, "Diabetes")
p3 = Patient(103,"nasir",17,"headach")
hospital.addpatient(p1)
hospital.addpatient(p2)
hospital.addpatient(p3)

# Adding Doctors
d1 = Doctor(201, "Dr. Ahmed", 45, "Cardiologist")
d2 = Doctor(202, "Dr. Khan", 38, "Physician")
hospital.adddoctor(d1)
hospital.adddoctor(d2)

# Scheduling Appointments
a1 = Appointment(p1, d1, "10-Jan-2026")
a2 = Appointment(p2, d2, "12-Jan-2026")
a3 = Appointment(p3,d1, "01-jan-2026")
hospital.scheduleappointment(a1)
hospital.scheduleappointment(a2)
hospital.scheduleappointment(a3)

# Billing
bill1 = Billing(p1, 3, 2000)
bill2 = Billing(p2, 5, 2500)
bill3 = Billing(p3,2, 2000 )

# Display Information
hospital.showpatients()
hospital.showdoctors()
hospital.showappointments()

print("\n--- Billing Details ---")
print(bill1)
print(bill2)
print(bill3)
```
