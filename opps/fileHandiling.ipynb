```python
#   for writing   
file= open("write_file.txt","w")  ##  "w"= mood
file.write("his is writing file")
file.close()
```

```python
#    for Reading   
file= open("write_file.txt","r")
print(file.read())
file.close()
```

```python
#   for adding   
file= open("write_file.txt","a")
file.write("this is new content")
file.close()
```
