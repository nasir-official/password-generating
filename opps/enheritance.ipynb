```python
#Single inheritance example
class Parent:
    def gree(self):
        print("Hello from Parent")
class Child(Parent):
    def greet(self):
        print("Hello from Child")
cal1 = Child()
cal1.greet()
cal1.gree()
```

```python
# multiple inheritance example
class Mother:
    def pro(self):
        print("mother class method")
class father:
    def skill(self):
        print("father class method")
class douther(Mother, father):
    def talent(self):
        print("douther class method  mother and father ")
d_obj = douther()
d_obj.pro()
d_obj.skill()
```

```python
# multi-level inheritance example
class GrandParent:
    def show1(self):
        print("GrandParent class")   
class Parent(GrandParent):
    def show2(self):
        print("Parent class")
class Child(Parent):
    def show3(self):
        print("Child class")
c_obj = Child()
c_obj.show1()
c_obj.show2()
c_obj.show3()
```

```python
# hierarchical inheritance example
class Parent:
    def display(self):
        print("Parent class method")    
class Child1(Parent):
    def show1(self):
        print("Child1 class method")
class Child2(Parent):
    def show2(self):
        print("Child2 class method")
class Child3(Parent):
    def show3(self):
        print("Child3 class method")
c1 = Child1()
c1.display()
```

```python
class iphone:
    def __init__(self,model,price):  #constructor
        self.model = model
        self.price = price
    def show(self):
        print(f"Model: {self.model}, Price: {self.price}")
    def purchase(self):
        print(f"Purchasing {self.model} for {self.price}")
class iphonepro(iphone):
    def __init__(self, model, price, camera): #constructor of child class
        super().__init__(model, price)  #for calling parent class constructor
        self.camera = camera
    def features(self):
        print("pro camera:", self.camera)
iphone1 = iphonepro("iphone14", 100000, "50MP")
iphone1.show()
iphone1.purchase()
iphone1.features()
```

```python
# hierarchical inheritance example
class Parent:
    def display(self):
        print("Parent class method")
class Child1(Parent):
    def show1(self):
        print("Child1 class method")
class Child2(Parent):
    def show2(self):
        print("Child2 class method")
class Child3(Parent):
    def show3(self):
        print("Child3 class method")
c1 = Child1()
c1.display()
```

```python
# hybrid inheritance example
class A:
    def method_a(self):
        print("Method A from class A")
class B(A):
    def method_b(self):
        print("Method B from class B")
class C(A):
    def method_c(self):
        print("Method C from class C")
class D(B, C):
    def method_d(self):
        print("Method D from class D")
d = D()
d.method_a()
```
