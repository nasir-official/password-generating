```python
#Single inheritance example
class Parent:
    def gree(self):
        print("Hello from Parent")
class Child(Parent):
    def greet(self):
        print("Hello from Child")
cal1 = Child()
cal1.greet()
cal1.gree()
```

```python
# multiple inheritance example
class Mother:
    def pro(self):
        print("mother class method")
class father:
    def skill(self):
        print("father class method")
class douther(Mother, father):
    def talent(self):
        print("douther class method  mother and father ")
d_obj = douther()
d_obj.pro()
d_obj.skill()
```

```python
# multi-level inheritance example
class GrandParent:
    def show1(self):
        print("GrandParent class")   
class Parent(GrandParent):
    def show2(self):
        print("Parent class")
class Child(Parent):
    def show3(self):
        print("Child class")
c_obj = Child()
c_obj.show1()
c_obj.show2()
c_obj.show3()
```
