```python
class Student:
    def __init__(self):
        self.name = "ali"
        self.father_name = "Muhammad ali"
        self.roll = 12
        self.age = 18

    def run_menu(self):
        while True:
            self.option = int(input("""Choice option
                               1= details
                               2= percentage
                               3 = grade
                               4=exit
                               """))
            if self.option == 1:
                self.detail()
            elif self.option == 2:
                self.per()
            elif self.option == 3:
                # Assuming you want to calculate grade based on percentage, but need perr
                # For now, prompt for percentage or integrate with per()
                perr = float(input("Enter percentage: "))
                self.grade(perr)
            elif self.option == 4:
                break
            else:
                print("Typing wrong")

    def detail(self):
        print(f"""
            Name: {self.name} 
            Father name: {self.father_name} 
            Roll_no : {self.roll}
            Age : {self.age}
         """)

    def per(self):
        marks = []
        for i in range(1, 5):
            try:
                a1 = int(input(f"Enter number of Book {i}  total marks is 50: "))
                marks.append(a1)
            except ValueError:
                print("Invalid input, enter a number.")
                return
        total = sum(marks)
        perr = total / 200 * 100
        print(perr)
        self.grade(perr)

    def grade(self, perr):
        if perr < 40:
            print("Try again")
        elif perr >= 40:
            print("pass")

user1 = Student()
user1.run_menu()
```
