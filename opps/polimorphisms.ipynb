```python
# polimorphisms in python
#overriding = different class same method ( with inheritance)
class Animal:
    def speak():
        print("Speak Animal")
class Child(Animal):
    def speak():
        print("Child Speaking")
s= Child()
```

```python
#Example 
class Cat:
    def sound(self):
        print("MMauooo")
class Dog:
    def sound(self):
        print("Brankkk")

def animal_info(animal): #for problem of overriding
    animal.sound()

animal_info(Cat()) #calling function bt parameter class 
animal_info(Dog())
```
