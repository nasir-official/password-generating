```python
#Operator Overloading  or  dunder function
class Student:
    def __init__(self,name,age,skills):
        self.name = name
        self.age = age
        self.skills = skills
    def __str__(self):
        return f"Name: {self.name}  |   Age: {self.age}  |  Skills: {self.skills})"
    def __len__(self):
        return len(self.skills)
    def __del__(self):
        print("Object Deleted")
    def __delitem__(self, index):
        del self.skills[index] 
info = Student("Ali", 20, ["Python", "Data Analysis", "Machine Learning"])
print(info)
print(len(info))
del info[1]
print(info.skills)
```
